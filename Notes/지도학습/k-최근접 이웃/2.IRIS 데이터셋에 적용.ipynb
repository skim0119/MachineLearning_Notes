## 데이터 불러오기

```python
# loading libraries
import pandas as pd

# define column names
names = ['sepal_length', 'sepal_width', 'petal_length', 'petal_width', 'class']

# loading training data
df = pd.read_csv('iris.data', header=None, names=names)
df.head()
```

## 데이터 살펴보기

```python
import matplotlib.pyplot as plt
import seaborn
seaborn.set(style='ticks')

classes = ['Iris-setosa','Iris-versicolor','Iris-virginica']

# Graph sepal length vs width
fg_sepal = seaborn.FacetGrid(data=df[['sepal_length','sepal_width','class']],
                       hue='class', hue_order=classes, aspect=1.68, size=6)
fg_sepal.map(plt.scatter, 'sepal_length', 'sepal_width').add_legend()

# Graph petal length vs width
fg_petal = seaborn.FacetGrid(data=df[['petal_length','petal_width','class']],
                       hue='class', hue_order=classes, aspect=1.68, size=6)
fg_petal.map(plt.scatter, 'petal_length', 'petal_width').add_legend()
```
