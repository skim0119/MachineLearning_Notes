# 수치형 데이터 정규화 (Normalization)

- 두가지 이상에 다른 속성값을 비교하거나 이용하기 위해 규칙을 가지고 데이터를 다듬는 과정

```python
# 예제 데이터
import numpy as np
from numpy.random import multivariate_normal, permutation
import pandas as pd
from pandas import DataFrame, Series
np.random.seed(100) # Fix

# Generate Dataset using multivariate_normal
n_category = 2
subdf_set = []
var = np.random.randint(7,15)
for category in range(n_category):
    mean = np.array([category*7])
    subdf = DataFrame(multivariate_normal(mean, np.eye(1)*var, 40))
    subdf_set.append(subdf)
df = pd.concat(subdf_set, ignore_index=True, axis=1)
df.plot()
```

## 평균/표준편차를 이용한 선형 방식
$$
x^* = \frac{x - \mu}{\sigma} \\
x \Rightarrow \text{original data}, \, x^* \Rightarrow \text{normalized data} \\
\mu \Rightarrow \text{mean}, \, \sigma \Rightarrow \text{standard deviation}
$$

```python
df_norm_MeanStd = (df-df.mean())/df.std()
df_norm_MeanStd.plot()
```

## 최대/최소를 이용한 선형 방식
$$
x^* = \frac{x - \min}{\max-\min} \\
x \Rightarrow \text{original data}, \, x^* \Rightarrow \text{normalized data}
$$

```python
df_norm_MinMax = (df-df.min())/(df.max()-df.min())
df_norm_MinMax.plot()
```
